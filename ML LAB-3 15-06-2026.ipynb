# Experiment 1: CIA Percentage Predicting GPA

```python
X = df[['Your CIA % of last semester']]
Y = df['Your GPA of last semester']
```

# Part C: Manual OLS Calculation
Task 17: Compute Slope and Intercept Manually

Formula:

b1​=∑(xi​−xˉ)2∑(xi​−xˉ)(yi​−yˉ​)​

b0​=yˉ​−b1​xˉ

```python
x = X.values.flatten()
y = Y.values

x_mean = np.mean(x)
y_mean = np.mean(y)

b1 = np.sum((x - x_mean)*(y - y_mean)) / np.sum((x - x_mean)**2)

b0 = y_mean - b1*x_mean

print("Manual Slope =", b1)
print("Manual Intercept =", b0)
```

Result

Manual slope and intercept calculated.

Explanation

Regression coefficients are calculated mathematically using OLS.

Task 18: Manual Prediction

```python
manual_predictions = b0 + b1 * X_test.values.flatten()

print(manual_predictions)
```

Result

Predicted GPA values generated manually.

# Comparison Task

# Parameter Saving Task
Task 20: Save Parameters Using Pickle

```python
parameters = {
    "slope": slope,
    "intercept": intercept
}

with open("linear_regression_weights.pkl", "wb") as file:
    pickle.dump(parameters, file)
```

Result

Parameters saved successfully.

Task 21: Load Pickle File

```python
with open("linear_regression_weights.pkl", "rb") as file:
    loaded_parameters = pickle.load(file)

print(loaded_parameters)
```

Result

Saved parameters loaded successfully.

Task 22: Prediction Using Loaded Parameters

```python
attendance = 90

predicted_gpa = (
    loaded_parameters["intercept"]
    + loaded_parameters["slope"] * attendance
)

print(predicted_gpa)
```

Result

Predicted GPA generated using loaded parameters.

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Independent and Dependent Variables
X = df[['Your CIA % of last semester']]
Y = df['Your GPA of last semester']

# Train Model
model = LinearRegression()
model.fit(X, Y)

# Predictions
y_pred = model.predict(X)

# Graph
plt.figure(figsize=(8,5))

plt.scatter(X, Y, label="Actual Data")

plt.plot(X, y_pred, color='red', linewidth=2,
         label="Regression Line")

plt.xlabel("CIA Percentage")
plt.ylabel("GPA")
plt.title("CIA Percentage vs GPA")
plt.legend()

plt.show()
```

# My Understanding 
This scatter plot reveals the relationship between a student's CIA (Continuous Internal Assessment) percentage and their GPA. The red regression line slopes upward from left to right, which clearly indicates a **positive correlation** between the two variables — meaning students who perform better in their internal assessments tend to achieve higher GPAs overall.

Looking at the data range, CIA percentages span from around 60% to 92%, while GPA values range from 3.0 to 3.8. Students with lower CIA scores in the 60–65% range mostly cluster around GPAs of 3.0 to 3.4, while students in the mid-range of 70–75% show a wider spread from 3.0 to 3.7, and those with higher CIA scores of 80–91% generally land between 3.4 and 3.8.

However, the data points are not tightly packed around the regression line — there is quite a bit of scatter. This tells us that while the trend is real, the correlation is **moderate rather than strong**. CIA performance alone does not fully determine a student's GPA, and other academic factors such as end-semester exams, project work, and attendance likely play an equally significant role.

In conclusion, this graph teaches us that putting consistent effort into internal assessments does positively impact your GPA, but it is not the only factor that matters. Think of CIA as a strong foundation — it gives you an edge, but your overall academic performance depends on multiple contributions working together.
